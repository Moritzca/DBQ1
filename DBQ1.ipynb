```python
import nltk
import pandas as pd
import csv
#nltk.download()
```

```python
ECQ1 = open("Earnings Call Q1.txt").read()
tokenizer = nltk.RegexpTokenizer(r"\w+")
word = tokenizer.tokenize(ECQ1)
```

```python
sec = []
for a in range(len(word)):
    sec.append(int(round(a/len(word)*4320)))
```

```python
price = pd.read_csv('DB trading.txt')
print(price)
```

```python
speech=pd.DataFrame({'Second': sec,'Word': word,})
print(speech)
```

```python
merged = pd.merge(speech,price,how="left",on="Second")
print(merged)
```

```python
merged[["Second","Volume"]].plot.line(figsize=(30, 10), subplots=True, x='Second', title="Trade Volume")
```

```python
merged[["Second","Price(x1000)"]].plot.line(figsize=(30, 10), subplots=True, x='Second', title="Price in €x1000")
```

```python
merged[["Second","Diff(in%)"]].plot.line(figsize=(30, 10), subplots=True, x='Second', title="Difference")
```
